# ***FACE DETECTION FOR SECURE OPERATIONS***

This model uses facial recognition to detect and identify faces for checking if the customer is the same as mentioned in order details for package delivery to avoid discrepancy


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
!pip install opencv-python
```

```python
img = plt.imread('i.jpeg')
```

```python
plt.imshow(img)
```

```python
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
```

```python
detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
all_faces = detector.detectMultiScale(img)
all_faces
```

```python
for one_face in all_faces:
    x, y, w, h = one_face
    img = cv2.rectangle(img, (x,y), (x+w, y+h), (0,0,0), 2)
```

```python
plt.imshow(img)
plt.show()
```

# ***OPTIMAL SITE SELECTION FOR INDUSTRIAL SETUP***
This model helps to locate the best suitable place to set up stores which helps in making work easier by locating all the near by customer and pitch up to the center location for a particular area.For example-If brand like dominos needs to startup with a new shop so this model will help to locate all the regular customer ordering pizza from zomato,swiggy,etc in that particular area and tells the best suited area to install a new shop.


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=500,n_features=2,centers=3,random_state=3)
```

```python
X.shape
```

```python
y.shape
```

```python
X[:5]
```

```python
y[:10]
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(X)
```

```python
locations=model.cluster_centers_
locations
```

```python
plt.scatter(X[:,0],X[:,1]) #show location of customers
plt.scatter(locations[:,0],locations[:,1],s=300,marker='*') #show the location of pizza parlour
plt.show()
```

```python
predicted_labels = model.labels_
predicted_labels
```

```python
plt.scatter(X[:,0],X[:,1],c=predicted_labels) #show location of customers
plt.scatter(locations[:,0],locations[:,1],s=300,marker='*') #show the location of pizza parlour
plt.show()
```
